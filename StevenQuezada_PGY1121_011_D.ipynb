```python

import datetime

entradas = [None] + [''] *100
precios = [0, 120000, 80000, 50000]
asistentes = []





def menu():
  print("______________________________________")
  print("             Creativos.cl             ")
  print("      Concierto VIP Michael Jam       ")
  print("--------------------------------------")
  print(" ")
  print("1- Comprar entradas.")
  print(" ")
  print("2- Mostrar ubicaciones disponibles.")
  print(" ")
  print("3- Ver listado de asistentes.")
  print(" ")
  print("4- Mostrar ganancias totales.")
  print(" ")
  print("5- Salir.")
  print("_____________________________________")



def comprar_entrada():
  cant_entradas = int(input("Cuantas entradas desea comprar? Ingrese la cantidad (Solo se admite un maximo de 3 entradas): "))
  if cant_entradas < 1 or cant_entradas > 3:
    print("La cantidad de entradas debe ser de minimo 1 entrada y maximo 3 entradas. ")
    return

  for e in range (cant_entradas):
    ubicacion= lugar_disponible()
    if ubicacion==-1:
      print("Lugar no disponible.")
      return
    run = input("Ingrese su RUN sin puntos ni guion: ")
    if not run.isdigit():
      print("El RUN ingresado no es valido.")
      return
    asistentes.append({"run:", run, "lugar: ", ubicacion})
    entradas[ubicacion] = run
  print("La compra se ah realizado con exito.")



def lugar_disponible():
  print("Lugares disponibles: ")
  print("---------------------")
  print("| Escenario |")
  print("| |")
  print("---------------------")
  for i in range (1, 101, 10):
    print(' '.join(f"{x:2}" if x != 0 else ' x' for x in range(i, i + 10)))

  while True:
        ubicacion = int(input("Ingrese el numero del lugar que desea: "))
        if ubicacion >= 1 and ubicacion <= 100:
          if entradas[ubicacion] == '':
              return ubicacion
          else:
            print("El lugar seleccionado no se encuentra disponible.")
        else:
          print("Ubicación inválida. Ingrese un número de asiento válido.")


def listado_asistentes():
  if len(asistentes) ==0:
    print("No existen asistentes registrados.")
    return
  print("Asistentes registrados: ")
  asistentes_ordenados = sorted(asistentes, key=lambda x: int(x["run"]))
  for asistente in asistentes_ordenados:
    print(f"RUN: {asistente['run']} Lugar escogido: {asistente['ubicacion']}")



def mostrar_ganancias():
    totales = [0, 0, 0, 0]
    for asistente in asistentes:
        ubicacion = asistente["ubicacion"]
        totales[ubicacion] += precios[ubicacion]

    print("----------------------------------")
    print("Tipo de entrada | Cantidad | Total")
    print("----------------------------------")
    for i in range(1, len(totales)):
        cantidad = totales.count(precios[i])
        total = precios[i] * cantidad
        print(f"{tipo_entrada(i)} {precios[i]:9d} | {cantidad:8d} | {total:5d}")
    print("--------------------------------")
    print(f"Total | {len(asistentes):8d} | {sum(totales):5d}")

def tipo_entrada(ubicacion):
    if ubicacion == 1:
        return "Platinum"
    elif ubicacion == 2:
        return "Gold"
    elif ubicacion == 3:
        return "Silver"

def salir(nombre, apellido):
    fecha_actual = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    print(f"\n¡Adios, {nombre} {apellido}!")
    print(f"Fecha y hora: {fecha_actual}")

def main():
    nombre = input("Ingrese su nombre: ")
    apellido = input("Ingrese su apellido: ")

    while True:
        menu()
        opcion = input("Ingrese una opción: ")

        if opcion == "1":
            comprar_entrada()
        elif opcion == "2":
            lugar_disponible()
        elif opcion == "3":
            listado_asistentes()
        elif opcion == "4":
            mostrar_ganancias()
        elif opcion == "5":
            salir(nombre, apellido)
            break
        else:
            print("Opción inválida. Por favor, seleccione algunas de las opciones disponibles.")

main()




```
